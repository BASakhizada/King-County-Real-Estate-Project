## Deriving New Columns

```python
df_house.head()
```

### deriving new column from grade as 'apartment_grade'

```python
df_house.loc[(df_house['grade'] >= 1) & (df_house['grade'] <= 3), 'apartment_grade'] = 'No_Construction_Design'
df_house.loc[(df_house['grade'] >= 4) & (df_house['grade'] <= 7), 'apartment_grade'] = 'Average_Construction_Design'
df_house.loc[(df_house['grade'] >= 8) & (df_house['grade'] <= 13), 'apartment_grade'] = 'High_Construction_Design'
```

```python
df_house.head()
```

### deriving variable from view column as house_view. 

```python
df_house.loc[df_house['view'] == 0, 'house_view'] = 'No_View'
df_house.loc[df_house['view'] == 1, 'house_view'] = 'Fair_View'
df_house.loc[df_house['view'] == 2, 'house_view'] = 'Average_View'
df_house.loc[df_house['view'] == 3, 'house_view'] = 'Good_View'
df_house.loc[df_house['view'] == 4, 'house_view'] = 'Excellent_View'
```

```python
df_house.head()
```

### deriving a column from condition column as apartment_condition

```python
df_house.loc[df_house['condition'] == 1, 'house_condition'] = 'Poor'
df_house.loc[df_house['condition'] == 2, 'house_condition'] = 'Fair'
df_house.loc[df_house['condition'] == 3, 'house_condition'] = 'Average'
df_house.loc[df_house['condition'] == 4, 'house_condition'] = 'Good'
df_house.loc[df_house['condition'] == 5, 'house_condition'] = 'Very Good'
```

```python
df_house.head()
```

### Creating an Histogram to see distribution of house prices 

```python
# Importing matplotlib.pyplot
import matplotlib.pyplot as plt
# Creating the histogram with 50 bins
hist_price = df_house['price'].plot.hist(bins =50)
plt.title('hist_price')

# Adding the title for the plot
plt.title('Price Distribution')

# Adding a label for the x-axis
plt.xlabel('Price in Million USD')

# Show the plot
plt.show()
```

As we see from the histogram, majority of the house prices fall between 0 - 1 million.A percentage of houses cost between 1 - 2 million and only a small number of houses cost more than 2 million USD.

### Scatter plot showing the relationship between price and the year houses built

```python
scatterplot_yr_built = sns.scatterplot(x = 'yr_built', y = 'price',data = df_house)
plt.title('scatterplot_yr_built')
```

There's no strong upward or downward trend between the year built and price. While we might expect newer homes to be more expensive, the prices appear fairly dispersed across all years, suggesting other factors are influencing price more strongly.A few houses have very high prices (close to 8 million), which stand out significantly from the rest of the data. These could represent luxury properties or houses with unique features, regardless of the year built. The majority of houses, regardless of when they were built, are clustered below 2 million in price. There seems to be a slight increase in the upper range of house prices for houses built after 1980 or 2000, but it's not a very strong pattern.

```python
# Set the figure size
plt.figure(figsize=(10,6))

# Plot the bubble chart
plt.scatter(x=df_house['sqft_living'], y=df_house['price'], 
            s=df_house['bedrooms'] * 50,  # Scale bedroom size to adjust bubble size
            alpha=0.1, c='blue')  # Set alpha for transparency and color

# Add labels and title
plt.title('Bubble Chart: Price vs Sqft Living with Bedrooms as Bubble Size')
plt.xlabel('Sqft Living')
plt.ylabel('Price')

# Display the plot
plt.show()
```

As the square footage of a house increases, its price also increases. The majority of houses with lower square footage tend to have lower prices, while homes with larger square footage (above 5,000 sqft) tend to have higher prices. Additionally, the size of the bubbles suggests that houses with more bedrooms also tend to have larger square footage. The smaller bubbles (representing fewer bedrooms) are clustered around the lower square footage, while larger bubbles (representing more bedrooms) appear as the square footage increases. This is logical because larger homes often have more bedrooms.
